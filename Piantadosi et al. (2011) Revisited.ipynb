# All Languages #

```r
allResults = list(frenchResults, spanishResults, englishResults, dutchResults,
italianResults, portugueseResults, germanResults, swedishResults, polishResults, romanianResults, czechResults)
allScores = do.call('rbind', lapply(allResults, function(x){x[['scores']]}))
allDFs = do.call('rbind', lapply(allResults, function(x){x[['df']]}))
```

```r
dictProps = do.call('rbind',lapply(split(allDFs, allDFs$language), function(df){
    inDictionary = (nrow(subset(df, group == 'in dictionary')) / nrow(df))
    outOfDictionary = (nrow(subset(df, group == 'not in dictionary')) / nrow(df))
    english = (nrow(subset(df, group == 'english')) / nrow(df))
    return(data.frame(language=df$language[1], group = c('In Dictionary', 'Not In Dictionary', 'English'
    ), proportion= c(inDictionary, outOfDictionary, english)))    
}))
dictProps$group = factor(dictProps$group, levels=c('English','Not In Dictionary', 'In Dictionary'))

propGroupsPlot = ggplot(dictProps) + geom_bar(aes(x=language, y=proportion, fill=group
), stat='identity') + ylab('Proportion') + xlab('Language') + theme(axis.text.x = element_text(angle = 45, hjust = 1))

pdf(paste0(plotPath,'propGroupsPlot.pdf'), width=8, height=6)
print(propGroupsPlot)
dev.off()
options(repr.plot.width=6, repr.plot.height=4, jupyter.plot_mimetypes = 'image/png')
propGroupsPlot
```

Frequency ~ word length correlation is higher and surprisal ~ word length correlation is lower when we subset to items in each dictionary. 

```r
allScores$language = as.factor(allScores$language)
allScores$language = with(allScores, factor(language, levels = rev(levels(language))))
```

```r
fourGroup_replicationPlot = ggplot(subset(allScores, (analysis %in% c('All types (replication)',
'English types','Types not in dictionary', 'Types in dictionary')))) + geom_bar(aes(x=language, 
y=cor, fill=cor_type), stat="identity", position="dodge")+ facet_wrap(~analysis, nrow= 1
) + ylab(expression(paste("Spearman's ", rho))) + xlab('Language') + theme(axis.text.x = element_text(angle = 90,
hjust = 1), strip.text = element_text(size=8)) + theme(legend.position="none") + coord_flip() + theme(
   axis.line = element_line(colour = "black"),panel.grid.major = element_blank(
   ),panel.grid.minor = element_blank(),panel.border = element_blank(),panel.background = element_blank(
   ), axis.text.x = element_text(colour = "black"),axis.text.y = element_text(
       colour = "black")) + geom_hline(yintercept=0, colour='black', linetype='dashed') + scale_fill_brewer(palette="Set1")

options(repr.plot.width=8, repr.plot.height=2.5,jupyter.plot_mimetypes = 'image/png')
fourGroup_replicationPlot

pdf(paste0(plotPath,'4groups_replicationPlot.pdf'), width=8, height=2.5)
fourGroup_replicationPlot
dev.off()
```

Within each group: 
* **Types in dictionary:** correlations with frequency and surprisal are both relatively high
* **Out-of-dictionary types:** Varies widely by language. This probably reflects differences in the number of unidecoded items that are not in the relevant dictionart (e.g. Spanish 'ninos'). Need to look at new analysis where UTF-8 characters are preserved to check this hypothesis. 
* **English types:** correlation with frequency is higher than surprisal
